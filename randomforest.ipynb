```python
#this are the selected parametrs from the actual data which is named as data 
selected_features = data[[
    'Target',
    'Blood Glucose Levels',
    'Insulin Levels',
    'BMI',
    'Age',
    'Blood Pressure',
    'Cholesterol Levels',
    'Waist Circumference',
    'Family History',
    'Genetic Markers',
    'Physical Activity',
    'Dietary Habits',
    'Smoking Status',
    'Alcohol Consumption',
    'Glucose Tolerance Test',
    'Pancreatic Health',
    'Liver Function Tests',
    'Early Onset Symptoms'
]]
selected_features.head()
```

```python
from sklearn.preprocessing import LabelEncoder

col_encode=[
    'Blood Glucose Levels',
    'Insulin Levels',
    'BMI',
    'Age',
    'Blood Pressure',
    'Cholesterol Levels',
    'Waist Circumference',
    'Family History',
    'Genetic Markers',
    'Physical Activity',
    'Dietary Habits',
    'Smoking Status',
    'Alcohol Consumption',
    'Glucose Tolerance Test',
    'Pancreatic Health',
    'Liver Function Tests',
    'Early Onset Symptoms'
]

le=LabelEncoder()
for col in col_encode:
    selected_features[col]=le.fit_transform(selected_features[col])

selected_features.head(10)
    
```

```python
selected_features.head(50)
```

```python
x=selected_features[['Blood Glucose Levels',
    'Insulin Levels',
    'BMI',
    'Age',
    'Blood Pressure',
    'Cholesterol Levels',
    'Waist Circumference',
    'Family History',
    'Genetic Markers',
    'Physical Activity',
    'Dietary Habits',
    'Smoking Status',
    'Alcohol Consumption',
    'Glucose Tolerance Test',
    'Pancreatic Health',
    'Liver Function Tests',
    'Early Onset Symptoms']]

y=selected_features['Target']  # This contains the actual diabetes type diagnosis
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=200,        # more trees for stability
    max_depth=15,            # limit tree depth
    min_samples_split=15,    # require more samples to split
    min_samples_leaf=5,      # require more samples at each leaf
    max_features='sqrt',     # random subset of features per split
    oob_score=True,          # enables internal validation
    random_state=42,
    n_jobs=-1           )   
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print('Confusion Matrix:',conf_matrix)  
```

```python
# Save the trained model and scaler properly
import pickle

print("Saving model and scaler...")

# Save the RandomForest model
with open('randomforest.pkl', 'wb') as f:
    pickle.dump(rf_model, f)
print("âœ“ Model saved to randomforest.pkl")

# Save the scaler
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
print("âœ“ Scaler saved to scaler.pkl")

# Verify the saved files
print("\nVerifying saved files...")
model_test = pickle.load(open('randomforest.pkl', 'rb'))
scaler_test = pickle.load(open('scaler.pkl', 'rb'))
print(f"Model type: {type(model_test)}")
print(f"Scaler type: {type(scaler_test)}")
print(f"Model expects {model_test.n_features_in_} features")
print(f"Number of classes: {len(model_test.classes_)}")
print(f"Sample classes: {list(model_test.classes_[:5])}")
```

```python
# ================================
# OVERFITTING CHECK
# ================================

print("="*70)
print("ðŸ” CHECKING FOR OVERFITTING")
print("="*70)

# 1. Calculate training accuracy
y_train_pred = rf_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

# 2. Compare with test accuracy
test_accuracy = accuracy_score(y_test, y_pred)

print(f"\nðŸ“Š ACCURACY COMPARISON:")
print(f"   Training Accuracy: {train_accuracy:.4f} ({train_accuracy*100:.2f}%)")
print(f"   Test Accuracy:     {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"   Difference:        {abs(train_accuracy - test_accuracy):.4f} ({abs(train_accuracy - test_accuracy)*100:.2f}%)")

# 3. Interpret results
print(f"\nðŸŽ¯ OVERFITTING ASSESSMENT:")
accuracy_gap = train_accuracy - test_accuracy

if accuracy_gap < 0.05:
    print("   âœ… NO OVERFITTING - Model generalizes well!")
    print("   The model performs similarly on both training and test data.")
elif accuracy_gap < 0.10:
    print("   âš ï¸ SLIGHT OVERFITTING - Acceptable for most cases")
    print("   Small gap between training and test accuracy.")
elif accuracy_gap < 0.15:
    print("   âš ï¸ MODERATE OVERFITTING - Consider regularization")
    print("   Noticeable gap between training and test accuracy.")
else:
    print("   âŒ SEVERE OVERFITTING - Model memorizing training data!")
    print("   Large gap indicates poor generalization.")

 
```
